Soit :

texte = "Python est puissant et Python est facile"
Compte le nombre de mots
Compte le nombre dâ€™occurrences du mot "Python"
Remplace "Python" par "Java"
Transforme la phrase en snake_case

```python


def analyse_texte(texte):
    nb_mot=len(texte.split())
    nb_occurence=texte.lower().count("python")
    remplace_texte=texte.replace("Python","java")
    return nb_mot,nb_occurence,remplace_texte
texte="Python est puissant et Python est facile"
result=analyse_texte(texte)
print(f"le nombre de mot est : {result[0]}")
print(f"le nombre d'occurence de python est : {result[1]}")
print(f"le mot python est remplace par: {result[2]}")
```
