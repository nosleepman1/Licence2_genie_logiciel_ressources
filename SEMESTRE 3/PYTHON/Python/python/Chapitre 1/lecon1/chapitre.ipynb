## liste de comprehension et utilisation de min max et sum
soit la liste suivante :
L=[3,10,0,5,11,4,8,16,17]
1. le carre de chaque element de la liste
2. le min et max
3. afficher la somme des nombre pairs
4. le max impair le plus grand
5. le nombre de nombre pair

```python
L=[3,10,0,5,11,4,8,16,17]
print(L)
# le carre de chaque element de la liste
liste_carre=[nbr**2 for nbr in L]
print(f"le carre de chaque element de la liste est : {liste_carre}")
print(f"le plus petit nombre de la liste est : {min(L)}")
print(f"le plus grand nombre de la liste est {max(L)}")
print(f"la somme des nombre de la liste est : {sum(L)}")
somme_paire=sum([nbr for nbr in L if nbr%2==0])
print(f"la somme des nombre pairs est : {somme_paire}")
max_impair=max([nbr for nbr in L if nbr%2!=0])
print(f"le nombre impair le plus grand est  : {max_impair}")
nbre_pair=sum(1 for nbr in L if nbr%2==0)
print(f"le nombre de nombre pairs est : {nbre_pair}")
```

## les fonctions simples


```python
def salutation(prenom,nom):
    return f"bonjour {prenom} {nom}"
salutation("astou","gassama")
```

```python
def f(x):
    return  x**2 - 2*x + 8
f(-1)
```

```python
def min_max(liste):
    plus_grand=max(liste)
    plus_petit=min(liste)
    return  plus_grand,plus_petit#ca retourne un tuple avec le plus petit et le plus grand
resultats=min_max([4,12,-3,0,11,-19,100])
print(resultats)
print(f"le plus grand nombre est : {resultats[0]}")
print(f"le plus petit nombre est : {resultats[1]}")
```
