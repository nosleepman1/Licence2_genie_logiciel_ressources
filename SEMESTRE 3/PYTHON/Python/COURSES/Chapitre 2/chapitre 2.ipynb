2.match case

```python
jour=input("entrer le jour : ").strip().lower()
match jour:
    case 'lundi':
        print("debut de la semaine")
    case 'vendredi':
        print("fin de la semaine")
    case 'samedi' | 'dimanche':
        print("week-end")
    case _:
        print("milieu de la semaine")
```

### 3. La boucle for

```python
#parcours avec range
l=[3,1,10,-1,11,18,19,29]
N=len(l)
for i in range(N):
    print(l[i])
```

## parcourt avec in

```python
l=[4,7,0,10,11]
for nbr in l:
    print(nbr)
```
